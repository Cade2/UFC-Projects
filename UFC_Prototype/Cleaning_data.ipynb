```python
missing_rows = fighter_tott[fighter_tott.isnull().any(axis=1)]
missing_rows
```

```python
stance_nan_values = fighter_tott['STANCE'].isna()

# Calculate the number of missing values for each stance based on the given percentages
total_nan = stance_nan_values.sum()
orthodox_count = int(total_nan * 0.76)
southpaw_count = int(total_nan * 0.21)
switch_count = total_nan - orthodox_count - southpaw_count

# Create an array with the respective counts for each stance
stance_fill_values = (
    ["Orthodox"] * orthodox_count +
    ["Southpaw"] * southpaw_count +
    ["Switch"] * switch_count
)

# Shuffle the array to randomize the assignment
np.random.shuffle(stance_fill_values)

# Fill the missing values in the 'STANCE' column with the generated values
fighter_tott.loc[stance_nan_values, 'STANCE'] = stance_fill_values


fighter_tott.isna().sum()
```
